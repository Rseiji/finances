# Stocks

```python
start_date = '2024-01-01'
end_date = '2024-12-31'

year_transactions = read_sql_query(
    f"""
    SELECT *
    FROM stocks.transactions
    WHERE DATE BETWEEN date '{start_date}' AND date '{end_date}'
    """
)
```

```python
year_transactions = year_transactions.sort_values(
    ["ticker", "date", "current_quantity"],
    ascending=[False, False, True]
)
```

```python
import pandas as pd

# Assuming `year_transactions` is your dataframe
def filter_last_date_rows(group):
    # Get the last date for the group
    last_date = group['date'].max()
    last_date_rows = group[group['date'] == last_date]

    if last_date_rows['quantity'].iloc[0] < 0:
        # For negative quantity, get the row with the smallest quantity
        return last_date_rows.loc[last_date_rows['current_quantity'].idxmin()]
    else:
        # For positive quantity, get the row with the greatest quantity
        return last_date_rows.loc[last_date_rows['current_quantity'].idxmax()]

# Apply the function to each group of 'ticker'
current_positions = (
    year_transactions.groupby('ticker', group_keys=False)
    .apply(filter_last_date_rows)
)
```

```python
current_positions
```

```python
current_positions["current_position"] = (
    current_positions["avg_price"]
    * current_positions["current_quantity"]
)
```

```python
year_paid_taxes = year_transactions.groupby("ticker", as_index=False)["taxes"].sum()
```

```python
year_profits = year_transactions.query("quantity < 0").assign(
    profit=lambda df: df["quantity"] * (-1) * (df["price"] - df["avg_price"]) - df["taxes"]
).groupby("ticker", as_index=False)["profit"].sum()
```

```python
year_report = current_positions.reset_index(drop=True)[
    ["date", "ticker", "current_quantity", "current_position"]
].merge(
    year_paid_taxes, how='outer', on='ticker'
).merge(
    year_profits, how='outer', on='ticker'
).fillna(0)
```
